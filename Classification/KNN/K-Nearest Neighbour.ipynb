## finding the best k value which suits for data

```python
training_acccuracy_result = np.empty(len(k))
test_accuracy_result = np.empty(len(k))


for i , j in enumerate(k):
    reg = KNeighborsClassifier(n_neighbors=j)
    reg.fit(x_train,y_train)
    training_acccuracy_result[i] = reg.score(x_train,y_train)
    test_accuracy_result[i] = reg.score(x_test,y_test)
    print(training_acccuracy_result[i] , test_accuracy_result[i])
    
```

## 3 type for Finding the k_value using cross validation technique 

```python
from sklearn.model_selection import cross_val_score
```

```python
c = []
for i in range(1,51):
    reg = KNeighborsClassifier(n_neighbors=i)
    s = cross_val_score(reg , x , y , cv=10)
    c.append(s.mean())
print(c)
```

```python
len(c)
```

```python
max(c)
```

```python
k
```

```python
for i , j in enumerate(c):
    if j == max(c):
        print(k[i])
    
```

```python
plt.figure(figsize = (15,5))

plt.plot(k , c , color = 'r' , marker = '.' ,linestyle = 'dashed' )

plt.show()
```

## from error point 

```python
c = []
for i in range(1,51):
    reg = KNeighborsClassifier(n_neighbors=i)
    s = cross_val_score(reg , x , y , cv=10)
    c.append(1 - s.mean())
print(c)
```

```python
len(c)
```

```python
min(c)
```

```python
k
```

```python
for i , j in enumerate(c):
    if j == min(c):
        print(k[i])
    
```

```python
plt.figure(figsize = (15,5))

plt.plot(k , c , color = 'r' , marker = '.' ,linestyle = 'dashed' )

plt.show()
```
