```python
import sklearn
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33,random_state=42)
```

```python
len(df)
```

```python
len(x_train),len(y_train)
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
reg=MultinomialNB()
```

```python
reg.fit(x_train,y_train)
```

```python
## checking model performance
## training data
```

```python
y_train_pred_na=reg.predict(x_train)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
```

```python
confusion_matrix(y_train,y_train_pred_na)
```

```python
228+8+36+109
```

```python
accuracy_score(y_train,y_train_pred_na)
```

```python
print(classification_report(y_train,y_train_pred_na))
```

```python
## for test data
```

```python
y_test_pred_na=reg.predict(x_test)
```

```python
confusion_matrix(y_test,y_test_pred_na)
```

```python
accuracy_score(y_test,y_test_pred_na)
```

```python
print(classification_report(y_test,y_test_pred_na))
```
